```python
import numpy as np # useful for manipulating matrices
from skimage import io, color, exposure, transform, filters, img_as_float # useful for image loading and saving
from scipy import ndimage 
from skimage.color import rgb2gray
from matplotlib import pyplot as plt # useful for plotting images on iPython or Jupyter Notebooks
```

```python
def gauss(sigma, SZ = 50):
    # LIVE CODE GAUSSIAN
    [xx,yy]= np.meshgrid(np.linspace(-4,4,SZ),np.linspace(-4,4,SZ))
    gaussian = np.exp(-0.5*(xx*xx+yy*yy)/(sigma*sigma))
    gaussian = gaussian/np.sum(gaussian)
    return gaussian

print(np.sum(gauss(1)))
plt.imshow(gauss(10), cmap = 'jet')
plt.colorbar()
```

```python
def inv_gaussian(sigma, SZ = 50):
    or_gauss = gauss(sigma, SZ)
    neg_gauss = -gauss(sigma, SZ)
    result = or_gauss.max() + neg_gauss
    result = result/np.sum(result)
    return result

print(np.sum(inv_gaussian(4)))
plt.imshow(inv_gaussian(5), cmap = 'jet')
plt.colorbar()
```

```python
import numpy as np
import scipy.stats as st

def gaussian2(nsig=3, kernlen=21):
    # STACK OVERFLOW GAUSSIAN WITH VARIABLE SIGMA AND SIZE

    x = np.linspace(-nsig, nsig, kernlen+1)
    kern1d = np.diff(st.norm.cdf(x))
    kern2d = np.outer(kern1d, kern1d)
    return kern2d/kern2d.sum()
```

```python
# image2 has to be bigger than image1
def hibridMaker(image1, image2):
    # Resize images, image1 will have the size of image2
    size1=int(round(image1.shape[0]*(4/3.))); # Columns
    size2=int(round(image1.shape[1]*(4/3.))); # Rows
    image1 = transform.resize(img1,(size1, size2));
    # CROP
    left = 50
    top = 50
    right = 51+image2.shape[0]-1
    bottom = 51+image2.shape[1]-1
    image1=image1[left: right, top:bottom]

    # Images to float, you have to give already loaded images
    image1 = img_as_float(image1)
    image2 = img_as_float(image2)
    
    # Declare low pass and highpass of images
    gaus = gauss(3, 51)
    lowPass1 = np.zeros((image1.shape[0], image1.shape[1], 3))
    highPass1 = np.zeros((image1.shape[0], image1.shape[1], 3))
    lowPass2 = np.zeros((image1.shape[0], image1.shape[1], 3))
    highPass2 = np.zeros((image1.shape[0], image1.shape[1], 3))
    # Convolve
    # Image 1
    lowPass1[:, :, 0] = ndimage.convolve(image1[:, :, 0], gaus)
    lowPass1[:, :, 1] = ndimage.convolve(image1[:, :, 1], gaus)
    lowPass1[:, :, 2] = ndimage.convolve(image1[:, :, 2], gaus)
    highPass1 = image1 - lowPass1
    # Image 2
    lowPass2[:, :, 0] = ndimage.convolve(image2[:, :, 0], gaus)
    lowPass2[:, :, 1] = ndimage.convolve(image2[:, :, 1], gaus)
    lowPass2[:, :, 2] = ndimage.convolve(image2[:, :, 2], gaus)
    highPass2 = image2 - lowPass2
    
    # Subplot time
    subplot, arr = plt.subplots(1, 4)
    
    arr[0].imshow(image1)
    arr[0].set_title("Image 1")
    arr[0].axis('off')
    
    arr[1].imshow(image2)
    arr[1].set_title("Image 2")
    arr[1].axis('off')
    
    temp1 = lowPass1 + highPass2
    arr[2].imshow(temp1, cmap = "gray")
    arr[2].set_title("Hibrid 1")
    arr[2].axis('off')
    
    temp2 = lowPass2 + highPass1
    arr[3].imshow(temp2, cmap = 'gray')
    arr[3].set_title("Hibrid 2")
    arr[3].axis('off')
    plt.gcf().set_size_inches((14,14))
    
#LOAD IMAGES
img1 = io.imread('human.png')
img2 = io.imread('cat.png')
hibridMaker(img1, img2)
```

```python
## Part 2
def findNextPowerOf2(n):
    k = 1
    while k < n:
        k = k << 1
    return k

from scipy import fftpack
def lowPass(image):
    # Get padding size
    img_shape = image.shape
    image = rgb2gray(image)
    frame = findNextPowerOf2(np.max(image.shape))
    # Fourier
    image = fftpack.fft2(image, (frame, frame))
    # Gaussian fourier
    SZ = 50
    gaus = fftpack.fft2(gauss(0.9, SZ), (frame, frame))
    #plt.imshow(np.log(1+np.abs(image)), cmap="jet") # Visualize image
    result = image * gaus
    # Crop image
    hs=int(np.floor(SZ/2.))
    result = fftpack.ifft2(result)
    im_crop = result[hs:img_shape[0]+hs, hs:img_shape[1]+hs]
    return im_crop

img2 = io.imread('human.png')
lp = lowPass(img2)
plt.imshow(np.real(lp), cmap = "gray")
plt.colorbar()
print(lp.shape)
```

```python
def highPass(imagen):
    # Cogemos la imagen LowPass
    LP = lowPass(imagen)
    # Creamos la matriz con el tamao de la imagen
    HP = np.zeros((LP.shape[0], LP.shape[1]))
    # Creamos la imagen highpass
    HP = rgb2gray(imagen) - LP
    # Devolvemos el highpass
    return HP

gato = io.imread('cat.png')
hp = highPass(gato)
plt.imshow(np.real(hp), cmap = "gray")
plt.colorbar()
print(hp.shape)
```
